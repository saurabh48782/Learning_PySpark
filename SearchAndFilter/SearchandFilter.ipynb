```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("SearchandFilter").getOrCreate()

spark
```

```python
path = '/Users/apple/Desktop/Learning_PySpark/Datasets/'
```

```python
fifa_data = spark.read.csv(path+'fifa19.csv', inferSchema=True, header=True)

fifa_data.limit(4).toPandas()
```

```python
fifa_data.printSchema()
```

```python
from pyspark.sql.functions import *
```

```python
# Using the Order by function to find out youngest and oldest football players with in the dataset

# Youngest player
fifa_data.select(['Name','Age']).orderBy(fifa_data['Age']).show(5)
```

```python
# Oldest player
fifa_data.select(['Name','Age']).orderBy(fifa_data['Age'].desc()).show(5)
```

```python
# Using like operator to filter out players who are associated with Barcelona football club
# Second False argument in the show() method helps prevent text wrapping in the output, if not passed long strings will be wrapped and not shown properly
fifa_data.select(['Name', 'Club']).where(fifa_data['Club'].like('%Barcelona%')).show(5, False)
```

```python
# To find substring from a string, we can use the substr() function
# The first argument in the substr() function is the starting index and second argument is the length of the substring that should be fetched
# Suppose, we want to check what is the file extension of all the photos, we can do that as:

fifa_data.select(['Photo',fifa_data.Photo.substr(-4,4)]).show(5, False)
```

```python
# Using the isin() function to check for existence of the arguments passed in the dataset
# Here, we want to fetch rows where the players are either from club FC Barcelona or Juventus

fifa_data[fifa_data['Club'].isin('FC Barcelona','Juventus', 'Real Madrid')].limit(5).toPandas()
```

```python
# Using the startswith() and endswith() function

# Suppose we want to filter out the name and club of footballers whose name starts with 'L' and endswith 'i'

fifa_data.select(['Name','Club']).where(fifa_data['Name'].startswith('L')).where(fifa_data['Name'].endswith('i')).show(5, False)
```

```python
# To slice rows from a dataframe, we can simply use the limit function

new_data = fifa_data.limit(100)

# Using the count method to verify only 100 records are fetched out of complete data
print(new_data.count())

new_data.limit(5).toPandas()
```

```python
# Taking out a subset of columns from the dataset
#col_list = ['Name', 'Age', 'Club', 'Nationality']

#new_data2 = fifa_data.select(col_list)

# we can fetch first five columns in the dataframe as
col_list = fifa_data.columns[:5]
new_data2 = fifa_data.select(col_list)

# using the len() function to find out length of columns list
print(len(new_data2.columns))

new_data2.limit(5).toPandas()
```

```python
# Creating a custom dataframe
custom_df = spark.createDataFrame([([1,2,3],),([4,5],)],['x'])
custom_df.show()
```

```python
# We can even slice arrays with in a column and it can be done using the slice() function
# The indexing in PySpark starts from 1 instead of 0 which is different from Python
# slice() function accepts three arguments, first one is the column name, second is the starting position and third is the length
custom_df.select(slice(custom_df.x,2,2)).show()
```

```python
# To change the naming convention we can use the alias() function
custom_df.select(slice(custom_df.x,2,2).alias('Col_X')).show()
```

```python
# Filtering out rows from the dataset
# We want to filter rows where players overall rating is greater than 90
fifa_data.filter("Overall>90").limit(5).toPandas()
```

```python
# If we want only few columns of information like Name, Age, Club and Overall rating for players whose overall rating is less than 50

fifa_data.filter("Overall<50").select(['Name','Age','Club', 'Overall']).limit(5).toPandas()
```

```python
# Collecting the results as an object type, so we can reference it later in our outputs
# Suppose, we just want to print the name of the player whose overall score is greater than 50 and he/she is the worst player

result = fifa_data.filter("Overall>50").select(['Name','Age','Club', 'Overall']).orderBy(fifa_data['Overall']).collect()

print("Worst player having overall score above 50 is: ", result[0][0])
```
